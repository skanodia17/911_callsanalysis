```python
# Add an Extra Reason column derived from an existing column
df['Reason']=df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()
```

```python
#Counplot of Reasons for calls
sns.countplot(x='Reason',data=df)
```

```python
#Check if the timestamps are objects/strings?
x=df['timeStamp'].iloc[0]
```

```python
type(x)
```

```python
#Convert them into Datetime object
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
df.head()
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['Hour']=df['timeStamp'].apply(lambda time: time.hour)
```

```python
df['Month']=df['timeStamp'].apply(lambda time: time.month)
```

```python
df['Day of Week']=df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df.head()
```

```python
dmap={0:'Mon',1:'Tues',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of Week']=df['Day of Week'].map(dmap)
```

```python
df.head()
```

```python
#Countplot of calls against the day of the week with reason as hues
sns.countplot(x='Day of Week',data=df,hue='Reason')
```

```python
df.head()
```

```python
sns.countplot(x='Month',data=df,hue='Reason')
```

```python
#Lineplot to draw b/w month against calls
byMonth=df.groupby('Month').count()
```

```python
byMonth.head()
```

```python
byMonth['lat'].plot()
```

```python
#Linear fit to calls per month against calls
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```
